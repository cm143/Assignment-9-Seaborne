```python
#exercise data
import seaborn as sns
import pandas as pd
exercise_data = pd.read_csv('Exercise_Data.csv')

#heatmap
numeric_data = exercise_data.drop(labels = ['id', 'diet', 'kind'], axis = 1)
numeric_data = (numeric_data - numeric_data.min()) / (numeric_data.max() - numeric_data.min())
sns.heatmap(numeric_data)
#categorical plot
sns.catplot(x = 'kind', y = 'diet', data = exercise_data)
#Create a categorical plot of pulse values by diet and by type of exercise
```

```python
#planet data
planet_data = sns.load_dataset('planets')
print(planet_data)

#relational plot
#scatterplot
sns.scatterplot(x = 'orbital_period', y = 'mass', data = planet_data)
```

```python
#lineplot
sns.lineplot(x = 'year', y = 'distance', data = planet_data)
```

```python
#histplot
sns.displot(x = 'mass', data = planet_data)
```

```python
#rugplot
sns.rugplot(x = 'orbital_period', hue = 'method', data = planet_data)
```

```python
#categorical plot
#violinplot
sns.violinplot(x = 'year', y = 'distance', data = planet_data)
```

```python
#barplot
sns.barplot(x = 'distance', y = 'method', data = planet_data)
```

I think that the scatterplot, displot, rugplot, and barplot all displayed something notable. The bar plot shows that the Microlensing method counted planets at a high distance, while the rug plot shows that the Imaging method counted high orbital periods. The scatter plot showed that a large number of planets have a low mass and low orbital period. The line graph showed that there were planets a far distance away between 2005 and 2010.
